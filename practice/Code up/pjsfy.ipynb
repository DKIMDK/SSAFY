```python
# swea 최빈값

def mode(lst):
    score_list = [0]*(max(lst)+1)
    max_frequency = 0
    max_mode = []
    for score in lst:
        score_list[score] += 1
        # print('score list : ',score_list)
        if score_list[score] > max_frequency:
            max_frequency = score_list[score]
            max_mode = [score]
            # print(max_mode)
        elif score_list[score] == max_frequency:
            max_mode.append(score)
            # print(max_mode)
    return max(max_mode)


test_cases = int(input())
for test_case in range(1,test_cases + 1):
    int(input())
    # num = test_case
    # print(num)
    score_list = list(map(int, input().split()))
    print(score_list)
    answer = mode(score_list)
    print(f'#{test_case} {answer}')
```

```python
#swea 두개의 숫자열
'''
N개의 숫자열 Ai와 M개의 숫자열 Bj가 있다.
두 숫자열이 서로 마주보게 자유롭게 위치 조정 후 마주보는 수끼리 곱해서 모두 더한 값의 최대값
단 더 긴 쪽의 양 끝을 벗어나서는 안됨.
'''
def solution(lst_a, lst_b):
    la = len(lst_a)
    lb = len(lst_b)

    if la == lb:
        answer = 0
        for i in range(lb):
            mul = lst_a[i]*lst_b[i]
            answer += mul
        return answer
    elif la > lb:
        window = 0
        answer = window 
        for i in range(0, la-lb+1):
            # print(f'{i} loop')
            window = 0
            for j in range(lb):
                window = window + lst_a[j+i]*lst_b[j]
                # print('window :', window)
            answer = max(answer, window)
            # print('answer : ',answer)
        return answer
    else:
        lst_a, lst_b = lst_b, lst_a
        return solution(lst_a, lst_b)

# 입력창
test_cases = int(input())
for test_case in range(1,test_cases + 1):
    len_a, len_b =map(int,(input().split()))  
    a = list(map(int, input().split()))
    # print(a)
    b = list(map(int, input().split()))
    # print(b)
    
    answer = solution(a, b)
    print(f'#{test_case} {answer}')
```
