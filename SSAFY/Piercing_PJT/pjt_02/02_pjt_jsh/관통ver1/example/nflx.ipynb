```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#A

df = pd.read_csv('../example/data/NFLX.csv', usecols = ['Date', 'Open', 'High', 'Low', 'Close'])
df.head()
```

```python
#B

df['Date'] = pd.to_datetime(df['Date'])
df = df[df['Date'] >= '2021-01-01']
plt.plot(df['Date'], df['Close'])
plt.title('NFLX Close Price')
plt.xlabel('Date')
plt.ylabel('Close')
plt.xticks(rotation = 45)
plt.show()
```

```python
#C

max_price = df['Close'].max()
min_price = df['Close'].min()

print(' :', max_price)
print(' :', min_price)
```

```python
#D

df['Month'] = df['Date'].dt.strftime('%Y-%m')
new_df = df[['Close','Month']].groupby('Month').mean().reset_index()
new_df
plt.plot(new_df['Month'], new_df['Close'])
plt.title('Monthly Average Close Price')
plt.xlabel('Date')
plt.ylabel('Average Close Price')
plt.xticks(rotation = 45)
plt.show()
```

```python
#E

df2 = df[df['Date']>= '2022-01-01']
plt.plot(df2['Date'], df2['High'], label = 'High')
plt.plot(df2['Date'], df2['Close'], label = 'Close')
plt.plot(df2['Date'], df2['Low'], label = 'Low')
plt.legend()
plt.title('High, Low and Close Prices since January 2022')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation = 45)
plt.show()
```
